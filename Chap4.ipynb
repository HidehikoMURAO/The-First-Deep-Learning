## 4.5 活性化関数

```python
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

### 4.5.1 ステップ関数

```python
import numpy as np
import matplotlib.pyplot as plt

def step_function(x): 
    return np.where(x <= 0, 0, 1)

x = np.linspace(-5, 5)
y = step_function(x)

plt.plot(x, y)
plt.show()
```

### 4.5.2 シグモイド関数

```python
import numpy as np
import matplotlib.pylab as plt

def sigmoid_function(x): 
    return 1 / (1 + np.exp(-x))

x = np.linspace(-5, 5)
y = sigmoid_function(x)

plt.plot(x, y)
plt.show()
```

### 4.5.3 tanh

```python
import numpy as np
import matplotlib.pylab as plt

def tanh_function(x): 
    return np.tanh(x)

x = np.linspace(-5, 5)
y = tanh_function(x)

plt.plot(x, y)
plt.show()
```

### 4.5.4 ReLU

```python
import numpy as np
import matplotlib.pylab as plt

def relu_function(x): 
    return np.where(x <= 0, 0, x)

x = np.linspace(-5, 5)
y = relu_function(x)

plt.plot(x, y)
plt.show()
```
